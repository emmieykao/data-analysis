#CATEGORY:

```python
categoryGroup = incidents_df.groupby(['category']).size()
categoryGroup.to_frame()
categoryGroup
```

```python
incidents_df.category.value_counts()
#puts it in order
```

```python
categoryGroup.plot(kind = 'bar')
```

Here, using the `incidents_df.groupby(['category']).size()` function, we can see the amount of crimes in each category. We can see that an overwhelming amount of crimes was in `DRUGS/ALCOHOL VIOLATIONS`, with $225707$ instances. This is one of the lower crime levels, usually being catagorized as an infraction or misdemeanor. The next largest groups of crimes was for `THEFT/LARCENY` with $134264$ instances and `VEHICLE BREAK-IN/THEFT` with $120952$ instances. These crimes can rank from low misdemeanors to high-level felonies.

---

The smallest amount of crimes took place under the `HOMICIDE` label. While only $487$ crimes were catagorized as a `HOMICIDE` ($\approx0.21\%$ of `DRUGS/ALCOHOL VIOLATIONS`), it can be inferred that this is the most extreme of all the crimes.

---

Therefore, it can be concluded that the most crimes took place under the labels of less extreme categories.

#TIME OF DAY:

```python
timeGroup = incidents_df.groupby(['is_night']).size()
timeGroup
```

```python
timeGroup.plot(kind = 'bar', title = 'Chart 1')
```

Here, we can see that there are more crimes committed under the `0` tag (crimes committed during the day) than those from dawn to dusk (the `1` tag). However, it is relatively close, so looking at a category-by-category representation of this data may be beneficial.

---

Still, we can conclude that on an overall basis, less crimes were committed during the night.

```python
allTimeGroups = incidents_df.groupby(['category', 'is_night']).size()
allTimeGroups = allTimeGroups.to_frame()
allTimeGroups
```

```python
allTimeGroups = allTimeGroups.unstack()
#allTimeGroups = allTimeGroups.transpose()
allTimeGroups.T
#what would be the optimal way to make this table? (I had to fiddle around with the .transpose() and .unstack() commands a few times)
```

```python
allTimeGroups.plot(kind = 'bar', title = 'Chart 2')
#is there I can color code this or make it more readable?
```

Here, the <font color='orange'>orange</font> bars represent the amount of crimes committed between dusk and dawn (at night), and the <font color='blue'>blue</font> bars are crimes committed during the day.

---

For several categories of crime, the quantities of nighttime- and daytime-crimes were relatively similar:

1. `ARSON`
1. `ASSAULT`
1. `DRUGS/ALCOHOL VIOLATIONS`
1. `HOMICIDE`
1. `MOTOR VEHICLE THEFT`
1. `ROBBERY`
1. `SEX CRIMES`
1. `VANDALISM`
1. `VEHICLE BREAK-IN/THEFT`
1. `WEAPONS`

Therefore, it can be concluded that for these ten categories of crime, there is a mostly equal percent chance of them occurring during the day or during the night.

---

There are also several categories in which a large majority of crimes occur during the day (similar to the trend shown in `Chart 1`):

1. `BURGLARY`
1. `FRAUD`
1. `THEFT`

This means that the police as well as the general public should be more wary of burglars, frauds, and thieves during the day, as there is a smaller probability of such crimes occuring at night.

---

Finally, a large amount of `DUI` (driving under the influence) violations took place at night. This makes sense, as many people tend to drive home from bars or parties after dusk. Police should be more aware of drunk drivers during the night, as a vast majority of people drive drunk between dusk and dawn.



#MONTH:

```python
monthGroups = incidents_df.groupby(['month']).size()
monthGroups = monthGroups.to_frame()
monthGroups
```

```python
monthGroups.plot(kind = 'bar', title = 'Chart 1')
```

Here, we can see that many crimes occur in the late spring to summer months, starting with April. From there, crime amounts steadily decrease and hit their low during November and February, with rates in January and December not much higher.

---

Therefore, we can conclude that police and others should be more wary during summer months.

```python
allMonthGroups = incidents_df.groupby(['category', 'month']).size()
allMonthGroups = allMonthGroups.to_frame()
allMonthGroups
```

```python
allMonthGroups = allMonthGroups.unstack()
allMonthGroups
```

```python
allMonthGroups.plot(kind = 'bar', title = 'Chart 2')
#how to fix key?
```

This graph actually gives a much clearer depiction of the month-by-month data than `Chart 1`, but it portrays similar trends. From the categories such as `DRUGS/ALCOHOL VIOLATIONS`, `THEFT/LARCENY`, and `VANDALISM`, we can see that spikes in crime occur either in early spring or mid-summer, so that is when authorities should be on the lookout for the most crime.

#DAY OF THE MONTH:

```python
dayGroup = incidents_df.groupby(['day']).size()
dayGroup = dayGroup.to_frame()
dayGroup
```

```python
dayGroup.plot(kind = 'bar', title = 'Chart 1')
```

Here, we can see that an overwhelming amount of crimes take place on the first day of a month, while the rest of the days seem to have relatively similar amounts of crime. Near the end, however, there does seem to be a slight drop in data, as the 29th and 30th of a month have noticeably less crime data. Finally, while the 31st seems to have a significantly low amounts of crime, only seven months have a 31st day.

```python
allDayGroups = incidents_df.groupby(['category', 'day']).size()
allDayGroups = allDayGroups.to_frame()
allDayGroups
```

```python
allDayGroups = allDayGroups.unstack()
allDayGroups
```

```python
allDayGroups.plot(kind = 'bar', title = 'Chart 2')
```

Similar to `Chart 1`, we can see that the vast majority of crimes occur at the beginning of the month. Specifically, `DRUGS/ALCOHOL VIOLATIONS` and `THEFT/LARCENY` have large peaks in the earlier months.

#DAY OF THE WEEK:

Here, we can see that many of the crimes took place later on during the week on Thursday or Friday (days `4` and `5`). However, not as many took place on the weekends (days `0` and `6`).

```python
allWeekDayGroups = allWeekDayGroups.unstack()
allWeekDayGroups
```

```python
allWeekDayGroups.plot(kind = 'bar', title = 'Chart 2')
```

Again, this chart shows similar trends to `Chart 1`. We can see that the vast majority of crimes occur late during the week. Specifically, `DRUGS/ALCOHOL VIOLATIONS` and `THEFT/LARCENY` are skewed left with peaks at days `4` and `5`.

While `Chart 1` seems to show a skewed right graph, the different categories vary in shapes and descriptions. `ASSAULT`, `BURGLARY`, and `DUI` (among others) are skewed right, following the trend of `Chart 1`. However, it is also important to look at the other categories, as those follow different patterns. It does appear that only `THEFT/LARCENY` has peaked most recently, so authorities should expect an overall decrease in crime.
